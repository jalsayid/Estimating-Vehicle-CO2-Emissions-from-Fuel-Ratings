```python
# To list the fields/variables, and identify their types
data = [
    {"Variable": "Year", "Data Type": "Categorical", "Dependency": "Independent", "Description": "the year the measurments were taken"},
    {"Variable": "Make", "Data Type": "Nominal", "Dependency": "Independent", "Description": "Manufacturer of the vehicle"},
    {"Variable": "Model", "Data Type": "Nominal", "Dependency": "Independent", "Description": "Model of the vehicle"},
    {"Variable": "Vehicle Class", "Data Type": "Nominal", "Dependency": "Independent", "Description": "Class of the vehicle"},
    {"Variable": "Engine Size (L)", "Data Type": "Numeric", "Dependency": "Independent", "Description": "Engine size in liters"},
    {"Variable": "Cylinders", "Data Type": "Numeric", "Dependency": "Independent", "Description": "Number of cylinders in the engine"},
    {"Variable": "Transmission", "Data Type": "Nominal", "Dependency": "Independent", "Description": "Type of transmission"},
    {"Variable": "Fuel Type", "Data Type": "Nominal", "Dependency": "Independent", "Description": "Type of fuel used"},
    {"Variable": "Fuel Consumption (L/100Km)", "Data Type": "Numeric", "Dependency": "Independent", "Description": "Fuel consumption rating in liters/100 km"},
    {"Variable": "Hwy (L/100 km)", "Data Type": "Numeric", "Dependency": "Independent", "Description": "Highway fuel consumption rating in liters/100 km"},
    {"Variable": "Comb (L/100 km)", "Data Type": "Numeric", "Dependency": "Independent", "Description": "Combined fuel consumption rating in liters/100 km"},
    {"Variable": "Comb (mpg)", "Data Type": "Numeric", "Dependency": "Independent", "Description": "Combined fuel consumption rating in mpg"},
    {"Variable": "CO2 Emissions (g/km)", "Data Type": "Numeric", "Dependency": "Dependent", "Description": "Tailpipe CO2 emissions in g/km"},
    {"Variable": "CO2 Rating", "Data Type": "Ordinal", "Dependency": "Independent", "Description": "CO2 emissions rating (1-10)"},
    {"Variable": "Smog Rating", "Data Type": "Ordinal", "Dependency": "Independent", "Description": "Smog emissions rating (1-10)"}
]

ndf = pd.DataFrame(data)
display(ndf)
```

```python
# In this step, we will carefully examine the dataset to identify any inconsistencies or instances of missing values.

## Note: as per required in the guidelines we have manipulated the data to demonstrate data cleaning. 

# Display info about the DataFrame
df_info = df.info()

# Check for columns with missing data
null_columns = df.columns[df.isna().any()]

# Printing columns names containing null values if any
if len(null_columns) > 0:
    print("\nColumns with missing data:")
    for i, col in enumerate(null_columns):
        if i < len(null_columns) - 1:
            print(col + ',', end=" ")
        else:
            print(col, end=" ")
    print()
else:
    print("\nThere are no columns with missing data.")
```

```python
##columns Make and Hwy (L/100km) has missing values, hence we will use the imputing approach since there is not alot of missing values.

#Since we have two variables with different data types:

#fill NaN values with mean or mode (modified)
for c in null_columns:
    if df[c].dtype!='object':
        value = df[c].mean()
    else:
        value = df[c].mode()
        value = value[0]  #0 will be the row(s) name
        
    df[c].fillna(value,inplace=True)

display(df.info()) #no null values !
```

```python
# Column Fuel Consumption (L/100km) should be classified as float and not object. 
# Based on the head() method a user entered not just the number but the unit as well in Fuel Consumption(L/100Km).
# Also we have noticed a negative values which is not possible.
# Hence, to resolve the inconsistency, we have used this approach:

# To remove the unit entered
df["Fuel Consumption (L/100Km)"] = df["Fuel Consumption (L/100Km)"].apply(
    lambda x: float(x.replace("L/100Km", "")) if type(x) == str else x)

 

# To remove negative values
seleted_cols = df.select_dtypes(exclude='object').columns
df[seleted_cols] = df[seleted_cols].applymap(lambda x: x if x >= 0 else abs(x))

display(df.info())
#display(df.head()) to double check!
```

```python
# The variable Year has only one value which is 2023, not helpful, we decided to drop the whole column.
df.drop("Year", axis=1, inplace=True)

# Also, we decided to drop the model column. Since, it has a lot of different values 
df.drop("Model", axis=1, inplace=True)

#display(df.head()) 
```

```python
# Now, we want to locate outliers and remove it:
## We will use boxplot to check outliers
import matplotlib.pyplot as plt
import seaborn as sns

# Identify numeric columns
num_columns = df.select_dtypes(exclude = 'object').columns

# Calculate the number of rows and columns for the grid
num_plots = len(num_columns)
num_rows = int(num_plots ** 0.5)
num_cols = (num_plots // num_rows) + (num_plots % num_rows > 0)

# Create the subplots
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 10))

# Flatten the axes array
axes = axes.flatten()

# Loop through each numeric column and create a boxplot
for i, (c, ax) in enumerate(zip(num_columns, axes)):
    sns.boxplot(y=c, data=df, ax=ax)
    ax.set_title(c)

# Adjust the layout and add a title to the overall figure
plt.tight_layout()
plt.suptitle('BoxPlots Before Removing Outliers', y=1.05)

# Display the plot
plt.show()
```

```python
# To find outliers using code 
columns_with_outliers = []

for col in num_columns:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1

    lower_whisker = q1 - 1.5 * iqr
    upper_whisker = q3 + 1.5 * iqr
    
    outliers_mask = (df[col] < lower_whisker) | (df[col] > upper_whisker)
    if any(outliers_mask):
        columns_with_outliers.append(col.capitalize())

if columns_with_outliers:
    columns = ', '.join(columns_with_outliers)
    print(f'There are outliers in the following numerical columns: {columns}')
    
else:
    print('No outliers found in any numerical columns.')
```

```python
# To remove the outliers:

# Identify numeric columns
num_columns = df.select_dtypes(exclude='object').columns

# Calculate the IQR range for each numeric column
Q1 = df[num_columns].quantile(0.25)
Q3 = df[num_columns].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds to identify outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers from each numeric column
for column in num_columns:
    df = df[(df[column] >= lower_bound[column]) & (df[column] <= upper_bound[column])]

# Draw the updated boxplot without outliers

# Identify numeric columns
num_columns = df.select_dtypes(exclude = 'object').columns

# Calculate the number of rows and columns for the grid
num_plots = len(num_columns)
num_rows = int(num_plots ** 0.5)
num_cols = (num_plots // num_rows) + (num_plots % num_rows > 0)

# Create the subplots
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 10))

# Flatten the axes array
axes = axes.flatten()

# Loop through each numeric column and create a boxplot
for i, (c, ax) in enumerate(zip(num_columns, axes)):
    sns.boxplot(y=c, data=df, ax=ax)
    ax.set_title(c)

# Adjust the layout and add a title to the overall figure
plt.tight_layout()
plt.suptitle('BoxPlots After Removing Outliers', y=1.05)

# Display the plot
plt.show()
```

# Phase 4: Model Building

In an era where environmental awareness is important, understanding and reducing the impact of vehicular CO2 emissions is a critical endeavor. This project addresses the pressing need for accurate prediction and classification of CO2 emissions based on a diverse set of automotive features. The ultimate goal is to develop a best-fit machine learning model that can effectively identify and classify the level of CO2 emissions associated with various vehicles.

---


***The reason why did we chose regression for our model:***
regression models were chosen for the CO2 emissions modeling due to the continuous numeric nature of the target variable "CO2 emissions". This approach allows us to predict and quantify emissions based on various independent automotive features. 

---

**Models Used:**
We have chosen three popular and effective regression models to test:
- Ordinary Least Squares
- Ridge
- Lasso

```python
import pandas as pd

ndf=df.drop(['Transmission','Make','Vehicle Class'],axis=1) # Drop irrelevant categorical features

# we will drop one of the 'Comb' columns since we have two columns with the same information but in different units 
#to avoid redundancy and simplify the analysis
ndf.drop(['Comb (mpg)'],axis=1,inplace=True)
display(ndf)
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder() # Initialize a LabelEncoder
ndf['Fuel Type'] = label_encoder.fit_transform(ndf['Fuel Type']) # to transform the 'Fuel Type' column from categorical to numerical as the model building requires numerical inputs

display(ndf)
```

```python
from sklearn.model_selection import train_test_split
import numpy as np


# 'X' contains the features, excluding the 'CO2 Emissions (g/km)' column
# 'y' contains the target variable, which is the 'CO2 Emissions (g/km)' column
X = ndf.drop(['CO2 Emissions (g/km)'],axis=1)
y = ndf['CO2 Emissions (g/km)']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=40)
```

```python
# Scaling the Train - Test splits
import numpy as np
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

scaler.fit(np.c_[X_train,y_train]) # Fit the scaler on the training data

# Transform the features and target variable in the X_train and y_train sets
A_train = scaler.transform(np.c_[X_train,y_train])
X_train = A_train[:,:-1]
y_train = A_train[:,-1]

# Transform the features and target variable in the X_test and y_test
A_test = scaler.transform(np.c_[X_test,y_test])
X_test = A_test[:,:-1]
y_test = A_test[:,-1]
```

```python
## OLS
from sklearn.linear_model import LinearRegression

# Initialize a LinearRegression and fit the model on the training data
reg1 = LinearRegression(fit_intercept=False).fit(X_train, y_train)
best_beta = np.round(reg1.coef_,2) #the best values for the coefficients

print(f'Trying OLS Regression:')
print(f'The best values for the estimates are :', best_beta.tolist())
```

```python
## Ridge
from sklearn.linear_model import RidgeCV

# Initialize a LinearRegression and fit the model on the training data
reg2 = RidgeCV(alphas=[1e-3, 1e-2, 1e-1, 1e0, 1e1, 1e2, 1e3], fit_intercept=False,cv=10).fit(X_train, y_train)
best_beta=np.round(reg2.coef_,2) #the best values for the coefficients

print(f'Trying Ridge Regression:')
print(f'The best values for the estimates are :', best_beta.tolist())
```

```python
## Lasso
from sklearn.linear_model import LassoCV

# Initialize a LinearRegression and fit the model on the training data
reg3 = LassoCV(alphas=[1e-3, 1e-2, 1e-1, 1e0, 1e1, 1e2, 1e3],fit_intercept=False,cv=10, random_state=0).fit(X_train, y_train)
best_beta=np.round(reg3.coef_,2)#the best values for the coefficients

print(f'Trying Lasso Regression:')
print(f'The best values for the estimates are :', best_beta.tolist())
```

#### Observations besed on the coefficients:
- OLS provides non-zero coefficients for most features
- Ridge Regression tends to shrink the coefficients due to the penalty term, and some are reduced compared to OLS.
- Lasso Regression tends to reduce some coefficients which are 'Engine Size (L)' and 'Fuel Type	' to exactly zero. It implies that these features with zero coefficients are considered less relevant or have a negligible impact on predicting the target variable.

```python
# Choosing the Best Model
# Regression Analysis: Mean Squared Error Metric
from sklearn.metrics import mean_squared_error

y_pred1 = reg1.predict(X_test) # Predictions using OLS
print('The MSE using OLS is:', mean_squared_error(y_test, y_pred1))

y_pred2 = reg2.predict(X_test) # Predictions using Ridge
print('The MSE using Ridge is:', mean_squared_error(y_test, y_pred2))

y_pred3 = reg3.predict(X_test) # Predictions using Lasso
print('The MSE using Lasso is:', mean_squared_error(y_test, y_pred3))
print('-'*80)
print("These values represent the average squared differences between the predicted and actual CO2 emissions. In this case, the Ridge Regression model has the lowest MSE, suggesting it performs slightly better on the test set compared to OLS and Lasso.")
```

### Findings:

The analysis of the data-driven CO2 emissions prediction revealed valuable insights about the performance of various models and affirmed the validity of our hypothesis through a comprehensive analysis of these models, where our hypothesis states that some of the input features provided in the dataset can be effectively utilized to predict the amount of CO2 emitted by a vehicle. Among the three models evaluated, the Ridge regression emerged as the most promising choice, showing the lowest Mean Squared Error of 0.03422681215002337.

```python
## Details of the best estimates
print('Ridge regression:')
print()
print('- The best penalty coefficient is:', reg2.alpha_)
print('- The best coefficient estimates are:', reg2.coef_.tolist())
print('- Mean Squared Error (MSE)=', mean_squared_error(y_test, y_pred2),'\n')
print("Interpretation of the coefficients:")
print('- Coefficients represent the impact of each predictor variable on CO2 emissions.')
print('- Positive coefficients suggest a positive relationship, while negative coefficients indicate a negative relationship.')
print('- The magnitude of coefficients reflects the strength of the impact.')
```
