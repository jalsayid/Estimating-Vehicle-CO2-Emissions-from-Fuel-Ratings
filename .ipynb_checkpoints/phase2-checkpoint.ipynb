```python
# To list the fields/variables, and identify their types
data = [
    {"Variable": "Year", "Data Type": "Categorical", "Dependency": "Independent", "Description": "the year the measurments were taken"},
    {"Variable": "Make", "Data Type": "Nominal", "Dependency": "Independent", "Description": "Manufacturer of the vehicle"},
    {"Variable": "Model", "Data Type": "Nominal", "Dependency": "Independent", "Description": "Model of the vehicle"},
    {"Variable": "Vehicle Class", "Data Type": "Nominal", "Dependency": "Independent", "Description": "Class of the vehicle"},
    {"Variable": "Engine Size (L)", "Data Type": "Numeric", "Dependency": "Independent", "Description": "Engine size in liters"},
    {"Variable": "Cylinders", "Data Type": "Numeric", "Dependency": "Independent", "Description": "Number of cylinders in the engine"},
    {"Variable": "Transmission", "Data Type": "Nominal", "Dependency": "Independent", "Description": "Type of transmission"},
    {"Variable": "Fuel Type", "Data Type": "Nominal", "Dependency": "Independent", "Description": "Type of fuel used"},
    {"Variable": "Fuel Consumption (L/100Km)", "Data Type": "Numeric", "Dependency": "Independent", "Description": "Fuel consumption rating in liters/100 km"},
    {"Variable": "Hwy (L/100 km)", "Data Type": "Numeric", "Dependency": "Independent", "Description": "Highway fuel consumption rating in liters/100 km"},
    {"Variable": "Comb (L/100 km)", "Data Type": "Numeric", "Dependency": "Independent", "Description": "Combined fuel consumption rating in liters/100 km"},
    {"Variable": "Comb (mpg)", "Data Type": "Numeric", "Dependency": "Independent", "Description": "Combined fuel consumption rating in mpg"},
    {"Variable": "CO2 Emissions (g/km)", "Data Type": "Numeric", "Dependency": "Dependent", "Description": "Tailpipe CO2 emissions in g/km"},
    {"Variable": "CO2 Rating", "Data Type": "Ordinal", "Dependency": "Independent", "Description": "CO2 emissions rating (1-10)"},
    {"Variable": "Smog Rating", "Data Type": "Ordinal", "Dependency": "Independent", "Description": "Smog emissions rating (1-10)"}
]

ndf = pd.DataFrame(data)
display(ndf)
```

```python
# In this step, we will carefully examine the dataset to identify any inconsistencies or instances of missing values.

## Note: as per required in the guidelines we have manipulated the data to demonstrate data cleaning. 

# Display info about the DataFrame
df_info = df.info()

# Check for columns with missing data
null_columns = df.columns[df.isna().any()]

# Printing columns names containing null values if any
if len(null_columns) > 0:
    print("\nColumns with missing data:")
    for i, col in enumerate(null_columns):
        if i < len(null_columns) - 1:
            print(col + ',', end=" ")
        else:
            print(col, end=" ")
    print()
else:
    print("\nThere are no columns with missing data.")
```

```python
##columns Make and Hwy (L/100km) has missing values, hence we will use the imputing approach since there is not alot of missing values.

#Since we have two variables with different data types:

#fill NaN values with mean or mode (modified)
for c in null_columns:
    if df[c].dtype!='object':
        value = df[c].mean()
    else:
        value = df[c].mode()
        value = value[0]  #0 will be the row(s) name
        
    df[c].fillna(value,inplace=True)

display(df.info()) #no null values !
```

```python
# Column Fuel Consumption (L/100km) should be classified as float and not object. 
# Based on the head() method a user entered not just the number but the unit as well in Fuel Consumption(L/100Km).
# Also we have noticed a negative values which is not possible.
# Hence, to resolve the inconsistency, we have used this approach:

# To remove the unit entered
df["Fuel Consumption (L/100Km)"] = df["Fuel Consumption (L/100Km)"].apply(
    lambda x: float(x.replace("L/100Km", "")) if type(x) == str else x)

 

# To remove negative values
seleted_cols = df.select_dtypes(exclude='object').columns
df[seleted_cols] = df[seleted_cols].applymap(lambda x: x if x >= 0 else abs(x))

display(df.info())
#display(df.head()) to double check!
```

```python
# The variable Year has only one value which is 2023, not helpful, we decided to drop the whole column.
df.drop("Year", axis=1, inplace=True)

# Also, we decided to drop the model column. Since, it has a lot of different values 
df.drop("Model", axis=1, inplace=True)

#display(df.head()) 
```

```python
# Now, we want to locate outliers and remove it:
## We will use boxplot to check outliers
import matplotlib.pyplot as plt
import seaborn as sns

# Identify numeric columns
num_columns = df.select_dtypes(exclude = 'object').columns

# Calculate the number of rows and columns for the grid
num_plots = len(num_columns)
num_rows = int(num_plots ** 0.5)
num_cols = (num_plots // num_rows) + (num_plots % num_rows > 0)

# Create the subplots
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 10))

# Flatten the axes array
axes = axes.flatten()

# Loop through each numeric column and create a boxplot
for i, (c, ax) in enumerate(zip(num_columns, axes)):
    sns.boxplot(y=c, data=df, ax=ax)
    ax.set_title(c)

# Adjust the layout and add a title to the overall figure
plt.tight_layout()
plt.suptitle('BoxPlots Before Removing Outliers', y=1.05)

# Display the plot
plt.show()
```

```python
# To find outliers using code 
columns_with_outliers = []

for col in num_columns:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1

    lower_whisker = q1 - 1.5 * iqr
    upper_whisker = q3 + 1.5 * iqr
    
    outliers_mask = (df[col] < lower_whisker) | (df[col] > upper_whisker)
    if any(outliers_mask):
        columns_with_outliers.append(col.capitalize())

if columns_with_outliers:
    columns = ', '.join(columns_with_outliers)
    print(f'There are outliers in the following numerical columns: {columns}')
    
else:
    print('No outliers found in any numerical columns.')
```

```python
#  هنا صارت لي مشكله😅مو كل ال اوتلايرز انشالو في عامودين ما نشال منهم كل شي مدري اذا عادي ولا لا
# To remove the outliers:

# Identify numeric columns
num_columns = df.select_dtypes(exclude='object').columns

# Calculate the IQR range for each numeric column
Q1 = df[num_columns].quantile(0.25)
Q3 = df[num_columns].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds to identify outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers from each numeric column
for column in num_columns:
    df = df[(df[column] >= lower_bound[column]) & (df[column] <= upper_bound[column])]

# Draw the updated boxplot without outliers

# Identify numeric columns
num_columns = df.select_dtypes(exclude = 'object').columns

# Calculate the number of rows and columns for the grid
num_plots = len(num_columns)
num_rows = int(num_plots ** 0.5)
num_cols = (num_plots // num_rows) + (num_plots % num_rows > 0)

# Create the subplots
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 10))

# Flatten the axes array
axes = axes.flatten()

# Loop through each numeric column and create a boxplot
for i, (c, ax) in enumerate(zip(num_columns, axes)):
    sns.boxplot(y=c, data=df, ax=ax)
    ax.set_title(c)

# Adjust the layout and add a title to the overall figure
plt.tight_layout()
plt.suptitle('BoxPlots After Removing Outliers', y=1.05)

# Display the plot
plt.show()
```
